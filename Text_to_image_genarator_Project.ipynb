```python
model_id1 = "dreamlike-art/dreamlike-diffusion-1.0"
model_id2 = "stabilityai/stable-diffusion-xl-base-1.0"


# Note: 'use_safetensors' is the correct argument name if you intend to use it, not 'use_satensors'.
# However, it's often not needed explicitly with from_pretrained as it's handled automatically.
# I'm removing the 'use_satensors' argument as it was flagged as unexpected.
pipe = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16, use_safetensors=True)
pipe = pipe.to("cpu")
```

```python
prompt = " a man is reading a book in a garden "
```

```python
image = pipe(prompt).images[0]
```

```python
print("[PROMPT]: ",  prompt)
plt.imshow(image)
plt.axis('off');
```

#Working with Stable Diffusion parameters

##Negative prompting
##num_inference_steps
##height
##weight
##num_images_per_prompt


```python
def generate_image(pipe,prompt,params):
  img = pipe(prompt, **params).images
  num_images = len (img)
  if num_images>1:
    fig, ax = plt.subplots(nrows=1,ncols=num_images)
    for i in range(num_images):
      ax[i].imshow(img[i])
      ax[i].axis('off')
  else:
    fig= plt.figure()
    plt.imshow(img[0]);
    plt.axis('off')
    plt.tight_layout()
```

```python
prompt = " give me one image girl face is very dirty face"
params = {}
```

```python
generate_image(pipe,prompt,params)
```
